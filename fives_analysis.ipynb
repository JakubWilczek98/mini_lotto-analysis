```python
# Modules
import pandas as pd
```

```python
# Load lotto data

data_file_names = ['mini-lottowyniki 2019',
                   'mini-lottowyniki 2020',
                   'mini-lottowyniki 2021',
                   'mini-lottowyniki 2022']

df_lotto_results = pd.DataFrame()

for name in data_file_names:
    df_lotto_results = df_lotto_results.append(pd.read_csv(f'data/{name}.csv', sep=';'), ignore_index=True)
```

```python
df_lotto_results.head(10)
```

```python
# Cutting off unnecessary data

df_lotto_results = df_lotto_results[['N1', 'N2', 'N3', 'N4', 'N5']]
```

```python
# Numbers are sorted, creating unique strings of five numbers

df_lotto_results['all_numbers'] = df_lotto_results.N1.astype(str) + ', ' +  df_lotto_results.N2.astype(str) + ', '\
+  df_lotto_results.N3.astype(str) + ', ' +  df_lotto_results.N4.astype(str) + ', ' +  df_lotto_results.N5.astype(str) 
df_lotto_results
```

```python
# Group by unique results and count them

df_counted_results = df_lotto_results.groupby(['all_numbers'])['all_numbers'].count()
```

```python
# Sort the results to find the most frequent fives 

df_counted_results.sort_values(ascending=False).head(150)
```
